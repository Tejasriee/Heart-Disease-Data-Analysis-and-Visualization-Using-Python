```python
#Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load Dataset
df = pd.read_csv('/content/drive/MyDrive/heart_disease1.csv')

print(df.head())



```

```python
#DATA CLEANING
print(df.isnull().sum())
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df.describe()
```

```python
#Average Values
avg = df[['Age',
          'BloodPressure',
          'Cholesterol',
          'HeartRate']].mean()

print(avg)

```

```python
def health_status(row):
      if row['BloodPressure'] > 145 or row['Cholesterol'] > 250:
          return "At Risk"
      else:
          return "Healthy"

df['Status'] = df.apply(health_status, axis=1)

print(df[['PatientID','Age','BloodPressure','Cholesterol','HeartRate','Status']])
```

```python
#Status Count
healthy = (df['Status'] == 'Healthy').sum()
risk = (df['Status'] == 'At Risk').sum()

print("Healthy Patients:", healthy)
print("At Risk Patients:", risk)
```

```python
#Status Graph
df['Status'].value_counts().plot(kind='bar')

plt.title('Patient Health Status')

plt.xlabel('Status')

plt.ylabel('Number of Patients')

plt.show()
```

```python
#Average Values Graph
avg.plot(kind='bar')

plt.title('Average Health Parameters')

plt.ylabel('Average Value')

plt.show()
```

```python
#Cholesterol Graph
plt.figure(figsize=(8,4))

plt.plot(df['Cholesterol'], marker='o')

plt.title('Cholesterol Levels')

plt.xlabel('Patient')

plt.ylabel('Cholesterol')

plt.show()
```

```python
#Blood Pressure Graph
plt.figure(figsize=(8,4))

plt.plot(df['BloodPressure'], marker='o')

plt.title('Blood Pressure Levels')

plt.xlabel('Patient')

plt.ylabel('Blood Pressure')

plt.show()
```

```python
#Heart Rate Graph
plt.figure(figsize=(8,4))

plt.plot(df['HeartRate'], marker='o')

plt.title('Heart Rate Levels')

plt.xlabel('Patient')

plt.ylabel('Heart Rate')

plt.show()
```

```python
#Scatter Plot (Blood Pressure vs Cholesterol)
plt.figure(figsize=(8,5))

plt.scatter(df['BloodPressure'], df['Cholesterol'])

plt.title('Blood Pressure vs Cholesterol')
plt.xlabel('Blood Pressure')
plt.ylabel('Cholesterol')

plt.show()

#If the points move upward together,it indicates Blood Pressure increases, Cholesterol also tends to increase.
#The scatter plot indicates a positive relationship between Blood Pressure and Cholesterol.
```

```python
#SCATTER PLOT (Age vs Blood Pressure)
plt.scatter(df['Age'], df['BloodPressure'])
plt.title("Age vs Blood Pressure")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.show()

#The scatter plot indicates a positive relationship between Blood Pressure and age.
```

```python
#Correlation Heatmap
plt.figure(figsize=(8,6))

sns.heatmap(
    df[['Age',
        'BloodPressure',
        'Cholesterol',
        'HeartRate']].corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title('Correlation Heatmap')

plt.show()
```
